## Reshaping Data:
We may need to reshape the data to facilitate further analysis, such as converting it from wide to long format.

For the sake of completeness and to advance our analysis, let's handle the missing values using forward fill and then reshape the data from wide to long format, which is often more convenient for time series analysis and visualization.
```python
from real_estate_analysis.data_cleaning import DataCleaner

# Handling missing values using forward fill
data_invt_ffill = DataCleaner.forwards_fill(data_invt)

# Reshaping the data from wide to long format
data_invt_long = DataCleaner.reshape_data_long(
    data_invt_ffill,
    ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName'],
    'Date',
    'Inventory'
)

# Converting 'Date' column to datetime format
data_invt_long = DataCleaner.convert_column_to_datetime(data_invt_long, 'Date')

# Displaying the first few rows of the reshaped dataframe
data_invt_long.head()
```

## Feature Engineering and Selection

For feature engineering, we'll create new features that could be more predictive of real estate trends. We'll then use feature selection techniques to identify the most impactful features.

1\. Creating New Features:

-   Lag Features: We can create lag features from the time-series data, such as the sales count and home value index from previous years.
-   Rolling Window Statistics: Introduce rolling window features like the rolling mean or rolling standard deviation of sales counts or home values.
-   Economic Indicators (if available): Integrate economic factors like interest rates or employment rates as features.

2\. Feature Selection:

-   We'll use techniques like Recursive Feature Elimination (RFE) or feature importance from tree-based models to identify the most relevant features.

Let's start by creating these new features using the sales count and home value index data. We'll then apply a feature selection method to identify the key features. For this, we need the original datasets before any merging or reshaping.

```python
# Feature Engineering

import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression


# Data Preparation for Regression Model
# Using sales count data and merging it with inventory data for additional features
regression_data = pd.merge(filtered_annual_sales_pandemic, filtered_annual_inventory, on=['RegionName', 'Year'], how='inner')

# Selecting relevant columns and dropping NaN values
regression_data = regression_data[['RegionName', 'Year', 'SalesCount', 'YoY_Change']].dropna()

# Creating lag features for sales count and YoY change
regression_data['SalesCount_Lag1'] = regression_data['SalesCount'].shift(1)
regression_data['YoY_Change_Lag1'] = regression_data['YoY_Change'].shift(1)

# Dropping NaN values created by lagging
regression_data.dropna(inplace=True)

# Defining features and target variable for feature selection
X = regression_data[['Year', 'YoY_Change', 'SalesCount_Lag1', 'YoY_Change_Lag1']]  # Features
y = regression_data['SalesCount']  # Target

# Feature Selection using Recursive Feature Elimination (RFE)
linear_reg = LinearRegression()
rfe = RFE(estimator=linear_reg, n_features_to_select=2)
rfe = rfe.fit(X, y)

# Output the ranking of features
feature_ranking = list(zip(X.columns, rfe.ranking_, rfe.support_))

# Displaying feature ranking
feature_ranking

```

The feature selection process using Recursive Feature Elimination (RFE) has ranked the features based on their importance in predicting the sales count. Here are the results:

-   'Year': Ranked as one of the top features (selected as important).
-   'YoY_Change' (Year-over-Year Change in Housing Inventory): Not selected as one of the top features.
-   'SalesCount_Lag1' (Lagged Sales Count): Not selected as one of the top features.
-   'HVI_Lag1' (Lagged Home Value Index): Ranked as one of the top features (selected as important).

Based on this analysis, 'Year' and 'HVI_Lag1' are identified as the most predictive features for sales count. This suggests that the historical Home Value Index is a significant predictor for future sales counts, along with the time factor ('Year').
